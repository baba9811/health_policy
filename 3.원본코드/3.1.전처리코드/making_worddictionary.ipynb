## 보건의료정책 전망과 과제 문서

## 두개의 문서를 하나로

```python
worddic = worddic1+worddic2
```

# 필요한 단어만 추출하여 단어사전 생성

```python
from konlpy.tag import Mecab
```

```python
def tokenizing(string):
    tokenizer = Mecab('C:\mecab\mecab-ko-dic')
    specific_tags = ['NNG', 'NNP', 'VV', 'VA'] # 일반명사, 고유명사, 동사, 형용사
    stopwords = []
    temp = []
    temp2 = []
    temp = tokenizer.pos(string) # pos tagging
    for j in range(len(temp)):
        if temp[j][1] in specific_tags:
            temp = [word for word in temp if not word in stopwords]
            temp2.append(temp[j][0])
    return set(temp2)
```

```python
word_set = tokenizing(worddic)
```

```python
word_dictionary = []
```

```python
for i in word_set:
    if len(i) > 1:
        word_dictionary.append(i)
```

## 단어사전을 pickle로 저장

```python
with open(f'word_dictionary.pickle',"wb") as f:
    pickle.dump(word_dictionary, f)
```
