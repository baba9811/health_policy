```python
from selenium import webdriver
import time
from tqdm import tqdm
import pandas as pd
```

```python
# 크롬 웹브라우저 실행
path = "chromedriver.exe"
```

#### 네이버 블로그는 573페이지까지 지원

```python
def blogcrawling(startdate, enddate):
    driver = webdriver.Chrome(path)
    url_list = []
    content_list = []
    text = "정신건강정책"
    
    for i in tqdm(range(1, 200)):  # range 안의 페이지의 블로그 내용을 읽어옴
        url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo='+ str(i) + '&rangeType=PERIOD&orderBy=recentdate&startDate='+startdate+'&endDate='+enddate+'&keyword=' + text
        driver.get(url)
        time.sleep(0.5)

        for j in range(1, 7): # 각 블로그 주소 저장
            titles = driver.find_element_by_xpath('/html/body/ui-view/div/main/div/div/section/div[2]/div['+str(j)+']/div/div[1]/div[1]/a[1]')
            title = titles.get_attribute('href')
            url_list.append(title)

    print("url 수집 끝, 해당 url 데이터 크롤링")

    for url in tqdm(url_list): # 수집한 url 만큼 반복
        driver.get(url) # 해당 url로 이동

        driver.switch_to.frame('mainFrame')
        overlays = ".se-component.se-text.se-l-default" # 내용 크롤링
        contents = driver.find_elements_by_css_selector(overlays)

        for content in contents:
            content_list.append(content.text)
    print('크롤링 종료')
    content_df = pd.DataFrame(content_list)
    content_df = content_df.drop_duplicates(keep='first',ignore_index=True)
    print(f'{startdate}~{enddate} 블로그 게시글 개수: ',len(content_df))
    content_df.to_csv('naverblog_{}_{}_{}.csv'.format(text,startdate,enddate),encoding='utf-8-sig',index=False)
```

```python
startdate = str(input('시작날짜 입력(yyyy-mm-dd): '))
print("시작날짜: ",startdate)
enddate = str(input('종료날짜 입력(yyyy-mm-dd): '))
print("종료날짜: ",enddate)
```

```python
blogcrawling(startdate, enddate)
```
